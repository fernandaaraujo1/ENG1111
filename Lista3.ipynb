```python
import pandas as pd
import os
from google.colab import drive

pd.set_option('display.max_columns',20)
drive.mount('/drive')
data = pd.read_csv("/drive/MyDrive/class_heart_disease.csv")
display(data[0:10])
```

```python
col_Thal = data["Thalach"]
col_Thal2=col_Thal.dropna()
display(col_Thal2[:10])
```

```python
media = col_Thal2.mean()
print(media)
```

```python
media = col_Thal2.mean()
coluna1=data.loc[(data['Thalach'])< media+5]
coluna2=coluna1.loc[(data['Thalach'])>media-5]
display(coluna2)
```

```python
apenas_5_col=coluna2[['Age','Sex','CP','Chol','Thalach']]
display(apenas_5_col)
```

```python
import random
novaColuna=(data['Age']-data['Sex'])*random.random()
print(novaColuna)
```

```python
import numpy as np
embaralhada=data.reindex(np.random.permutation(data.index))
pd.set_option('display.max_columns', 7)
pd.set_option('display.max_rows', 5)
display(embaralhada)
```

```python
data = data.sort_values(by='Age', ascending=True)
print(f"As menores 5 idades  : {data['Age'].iloc[0:5]}")
      
pd.set_option('display.max_columns', 7)
pd.set_option('display.max_rows', 5)

display(data)
```

```python
pd.set_option('display.max_columns', 7)
pd.set_option('display.max_rows', 15)


data = data.groupby('Age')[['Sex']].mean()

display(data)
```
